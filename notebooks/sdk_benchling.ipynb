```python
output_dir = '../output'
output_path = os.path.join(output_dir, "benchling_export.csv")
```

```python
# Retrieve mixtures
example = benchling.mixtures.list()
mixture = next(example, None)
print(mixture)
print(type(mixture))
```

```python
# Export data
os.makedirs(os.path.dirname(output_path), exist_ok = True)

with open(output_path, "w") as out:
    for item in mixture:
        out.write(str(item)+"\n")
```
