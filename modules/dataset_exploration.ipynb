```python
import os
import json
import pandas as pd
from pandas import json_normalize
```

```python
# set parameters and paths
data_dir = "../Data"
out_dir = "../outputs"
```

```python
# Import dataset
data_path = os.path.join(data_dir, "CrashMS_high resolution HD experiment.json")

with open(data_path, 'r') as data_file:
    data = json.load(data_file)
```

```python
norm_df = pd.json_normalize(data)
print(norm_df)
```

```python
# Explore first batch
print(norm_df.loc[[0],['batches']])
```

```python
# Explore all batches(23) with explode()
exploded_df = norm_df.explode("batches")
print(exploded_df)
```

```python
# Batches
batches = exploded_df.loc[0, "batches"]
```

```python
# Get common variables:
# Loop through each batch's variables
variables_sets = []
for batch in batches:
    variables = batch["variables"]
    variable_names = set(var["name"] for var in variables)  # Extract variable name
    variables_sets.append(variable_names)  # Add name to the list

# Intersect same names in all batches
common_variables = set.intersection(*variables_sets)

# Common variables
print(len(common_variables))
for variable in common_variables:
    print(variable)
```

```python
def explore(data):
    """
    This function explore general data from dataset project
    
    Parameters:
    data (obj): The dataset to explore

    Returns:
    list: List of project info
    
    """
    # Define data
    project_number = data["project"]
    project_id = data["id"]
    batches_number = len(data["batches"])
    batch_var_list = batch_variables(data)

    # Create list: general info
    project_info = [
        f"Project nÂº: {project_number}",
        f"Project ID: {project_id}",
        "----",
        f"Batches nÂº: {batches_number}",
        "-Variables per Batch:",
        *batch_var_list,
    ]

    # Return general info
    general_info = print('\n'.join(project_info))
    return general_info
```

```python
def batch_variables(data):
    """
    This function explores variables in each batch
    Creates a dictionary and prints them as a list
    """
    batch_var_dict = {}
    batch_var_list = []
    for index, batch in enumerate(data["batches"]):
        batch_var_dict[index] = len(batch['variables'])

    for index,var_count in batch_var_dict.items():
        batch_var_list.append(f"Batch {index}: {var_count} variables")
        
    return batch_var_list
```
